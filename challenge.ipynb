```python
# imports
import os
import torch
import torch.nn as nn
import numpy as np
import scipy.io
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
from collections import Counter
from PIL import Image
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report, confusion_matrix
from torch.utils.data import DataLoader, Dataset, WeightedRandomSampler
from torch.optim.lr_scheduler import ExponentialLR
from torchvision import transforms
from torchvision.models import resnet50
warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
# setup device for inference
device = torch.device("cuda") if torch.cuda.is_available() else torch.device("cpu")
print(f'Using {device} for inference')
```

```python
# Create a nn module using pretrained ResNet50 for the Flowers dataset
class FlowersResNet50(nn.Module):
    def __init__(self, num_classes):
        super(FlowersResNet50, self).__init__()
        self.model = resnet50(pretrained=True)
        num_ftrs = self.model.fc.in_features
        self.model.fc = nn.Sequential(
            nn.Dropout(0.5),  # Add dropout for regularization
            nn.Linear(num_ftrs, num_classes)  # Modify the final layer to match the number of classes
        )

    def forward(self, x):
        return self.model(x)
```

```python
# custom dataset class for the flowers dataset
class FlowersDataset(Dataset):
    def __init__(self, image_folder, labels_file, indices, transform=None):
        self.image_folder = image_folder
        self.transform = transform

        self.labels = scipy.io.loadmat(labels_file)['labels'].squeeze()
        self.image_files = os.listdir(image_folder)
        self.indices = indices
    
    def __len__(self):
        return len(self.indices)

    def __getitem__(self, index):
        real_idx = self.indices[index]
        img_name = os.path.join(self.image_folder, self.image_files[real_idx])
        image = Image.open(img_name).convert('RGB')

        if self.transform:
            image = self.transform(image)
        
        return image, self.labels[real_idx]
```

```python
# check the class distribution of the data
labels_file = "./imagelabels.mat"
labels_data = scipy.io.loadmat(labels_file)
labels = labels_data['labels'].squeeze()

class_counts = Counter(labels)

# Print the counts
print("Class distribution:")
for label, count in class_counts.items():
    print(f"Class {label}: {count} samples")

# Plot the class distribution
plt.bar(class_counts.keys(), class_counts.values())
plt.xlabel('Class')
plt.ylabel('Number of samples')
plt.title('Class distribution')
plt.show()
```

```python
# get the data split indices
data_split = scipy.io.loadmat('setid.mat')

# get the indices for training, testing, and validtaion data
train_idx = data_split['trnid'].squeeze()
test_idx = data_split['tstid'].squeeze()
val_idx = data_split['valid'].squeeze()
```

```python
# hyperparamters
LEARNING_RATE = 0.1
NUM_EPOCHS = 15
GAMMA = 0.9
```

```python
# training data transform - includes transforms to minimize overfitting
data_transform_train = transforms.Compose([
  transforms.RandomResizedCrop(224),
  transforms.RandomHorizontalFlip(),
  transforms.ColorJitter(),
  transforms.RandomRotation(10),
  transforms.ToTensor(),
  transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
])

# validation data transforms
data_transform_val = transforms.Compose([
  transforms.Resize(256),
  transforms.CenterCrop(224),
  transforms.ToTensor(),
  transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
])
images_dir = "./Flowers/jpg"

# store the training and validation datasets into memory (based on indices obtained from a previous step)
train_dataset = FlowersDataset(images_dir, labels_file, train_idx, transform=data_transform_train)
val_dataset = FlowersDataset(images_dir, labels_file, val_idx, transform=data_transform_val)

# method that provides weights for each of the labels in the dataset
def get_class_weights(labels):
    class_sample_count = np.array([len(np.where(labels == t)[0]) for t in np.unique(labels)])
    weight = 1. / class_sample_count
    samples_weight = np.array([weight[t - 1] for t in labels])
    return samples_weight

# class weighting to accomodate for the uneven class distribution
train_labels = np.array([label for _, label in train_dataset])
samples_weight = get_class_weights(train_labels)
samples_weight = torch.from_numpy(samples_weight).double()
sampler = WeightedRandomSampler(samples_weight, len(samples_weight))

# create data loaders from the training and validation dataset
train_data_loader = DataLoader(train_dataset, batch_size=32, sampler=sampler)
val_data_loader = DataLoader(val_dataset, batch_size=32, shuffle=True)

# get the total number of class labels
num_classes = len(class_counts)

# initialize the resenet50 model made for the flowers dataset
resnet50_model = FlowersResNet50(num_classes)

# loss function
loss_fuction = torch.nn.CrossEntropyLoss()

# optimizer
optimizer = torch.optim.SGD(resnet50_model.parameters(), lr=LEARNING_RATE)

# lr scheduler
scheduler = ExponentialLR(optimizer, gamma=GAMMA)
resnet50_model = resnet50_model.to(device)
```

```python
best_val_loss = float('inf')

# training process
for epoch in range(NUM_EPOCHS):
    print(f"Epoch {epoch+1}/{NUM_EPOCHS}")
    print("-" * 10)

    # set model to training phase
    resnet50_model.train()

    running_loss = 0.0
    running_corrects = 0

    # iterate over the data from the training data loader
    for inputs, labels in train_data_loader:
        inputs, labels = inputs.to(device), (labels - 1).to(device)

        optimizer.zero_grad()

        outputs = resnet50_model(inputs)

        _, preds = torch.max(outputs, 1)
        loss = loss_fuction(outputs, labels)

        loss.backward()
        optimizer.step()

        running_loss += loss.item() * inputs.size(0)
        running_corrects += torch.sum(preds == labels.data)

    # step the exponential LR scheduler
    scheduler.step()

    # calculate statistics
    epoch_loss = running_loss / len(train_data_loader.dataset)
    epoch_acc = 100. * running_corrects.double() / len(train_data_loader.dataset)

    print(f'Training    ==> Loss: {epoch_loss:.4f} | Acc: {epoch_acc:.2f}%')

    # Validation loop
    # set the model to validation phase
    resnet50_model.eval()
    val_loss = 0.0
    val_correct = 0
    val_total = 0
    with torch.no_grad():
        # iterate over the data from the validation data loader
        for inputs, labels in val_data_loader:
            inputs, labels = inputs.to(device), (labels - 1).to(device)
            outputs = resnet50_model(inputs)

            loss = loss_fuction(outputs, labels)
            val_loss += loss.item()
            _, preds = outputs.max(1)
            val_total += labels.size(0)
            val_correct += preds.eq(labels).sum().item()

    # calculate the statistics
    val_loss /= len(val_data_loader)
    val_acc = 100. * val_correct / val_total

    print(f'Validation  ==> Loss: {val_loss:.4f} | Acc: {val_acc:.2f}%\n')

    # save the best model to the best_model.pth file
    if val_loss < best_val_loss:
        best_val_loss = val_loss
        # save the model with the best loss value
        torch.save(resnet50_model.state_dict(), 'best_model.pth')
```

```python
# method to use the model with the validation dataset and return the labels vs predictions
def evaluate_model(model, dataloader, device):
    model.eval()  # Set model to evaluate mode
    all_labels = []
    all_preds = []

    with torch.no_grad():
        for inputs, labels in dataloader:
            inputs, labels = inputs.to(device), (labels - 1).to(device)
            outputs = model(inputs)
            _, preds = torch.max(outputs, 1)
            all_labels.extend(labels.cpu().numpy())
            all_preds.extend(preds.cpu().numpy())

    return np.array(all_labels), np.array(all_preds)

# instatiate and load the model with the saved best weights
model = FlowersResNet50(num_classes)
model_state_dict = torch.load("./best_model.pth")
model.load_state_dict(model_state_dict)

# data loader without shuffle (easier to calculate statistics)
val_data_loader2 = DataLoader(val_dataset, batch_size=32, shuffle=False)

# Get validation data
val_labels, val_preds = evaluate_model(model, val_data_loader2, device)

# Calculate metrics
accuracy = accuracy_score(val_labels, val_preds)
precision = precision_score(val_labels, val_preds, average='weighted')
recall = recall_score(val_labels, val_preds, average='weighted')
f1 = f1_score(val_labels, val_preds, average='weighted')

print('Validation Accuracy: {:.4f}'.format(accuracy))
print('Validation Precision: {:.4f}'.format(precision))
print('Validation Recall: {:.4f}'.format(recall))
print('Validation F1 Score: {:.4f}'.format(f1))
```

```python
# Detailed classification report
print('\nClassification Report:\n')
print(classification_report(val_labels, val_preds))
```

```python
# Confusion matrix
conf_matrix = confusion_matrix(val_labels, val_preds)
plt.figure(figsize=(25, 20))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=range(num_classes), yticklabels=range(num_classes))
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix')
plt.show()

"""The Dark Blue spots on the confusion matrix spots indicate where most of the predictions lie. Since, we can see that a majority of them appeared linearly with the true values of the labels, we can see that the model predictions were almost accuract with the true label values of the validation dataset"""
```
